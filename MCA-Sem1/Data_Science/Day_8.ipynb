```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
import seaborn as sb
```

# Import the CSV File

```python
df=pd.read_csv('USA_Housing.csv')
df
```

# Plot: Avg. Area Income vs. Price

```python
df_Price_Income=pd.DataFrame({'Price':round(df['Price'],3),'Income':round(df['Avg. Area Income'],3)})
df_Price_Income=df_Price_Income.sort_values(by='Income')
mpl.plot(df_Price_Income['Income'],round(df_Price_Income['Price'],3),'r')
mpl.xlabel('Income')
mpl.ylabel('Price')
```
